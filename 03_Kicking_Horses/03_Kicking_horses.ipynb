### 2. Kicking horses

A famous early  application of low-number statistics was an analysis of Prussian cavalryman horse-kick deaths by [Bortkiewicz](https://www.wikiwand.com/en/Ladislaus_Bortkiewicz) in 1898.

He studied the distribution of 122 men kicked to death by horses among 10 Prussian army corps within 20 years (so a total number of corpes of 200). He recorded the number of corps with a given number of deaths:

| Number of deaths | Number of groups |
| --- | --- |
| 0 | 109 |
| 1 | 65 |
| 2 | 22 |
| 3 | 3 |
| 4 | 1 |

- How does it look like? Again try to fit it by hand.

```python
import numpy as np
import pylab  as plt
from scipy.stats import poisson
```

```python
tot = 200
list_D = []
list_G = []

with open("Death_by_horse.txt", "r") as file:
    for line in file:
        value = line.split()  
        list_D.append(int(value[0]))  
        list_G.append(int(value[1]))  

DEATHS = np.array(list_D) 
GROUPS = np.array(list_G) / tot
```

```python
plt.scatter(DEATHS, GROUPS, color='red') 
plt.title('Number of Prussian cavalryman horse-kick deaths ')
plt.xlabel('Deaths in group')
plt.ylabel('Groups')

for mu in [0.1,0.5,1,2]:
    p = poisson.pmf(DEATHS, mu)
    plt.plot(DEATHS, p, label=r'$\mu='+str(mu)+"$" )
plt.legend()
plt.show()

mean = np.average(DEATHS, weights=GROUPS) 
print(f"Estimated mean of deaths: {mean}")
```

```python
plt.scatter(DEATHS, GROUPS, color='red') 
plt.title('Number of Prussian cavalryman horse-kick deaths - Poissonian Fit')
plt.xlabel('Deaths in group')
plt.ylabel('Groups')

p = poisson.pmf(DEATHS, mean)
plt.plot(DEATHS, p, label=r'$\mu='+str(mean)+"$", color = 'olive')

plt.bar(DEATHS, p, color = 'olive', alpha = 0.5)

plt.legend()
plt.show()
```
